```python
#impute object with mode - fill NaN
loan['Gender'].fillna('Male',inplace=True)
loan['Married'].fillna('Yes',inplace=True)
loan['Dependents'].fillna('0',inplace=True)
loan['Self_Employed'].fillna('No',inplace=True)
loan['Credit_History'].fillna('1.0',inplace=True)
loan.isnull().sum()
```

```python
#impute int(float) with mean
loan.fillna(loan['LoanAmount'].mean(),inplace=True)
loan.fillna(loan['Loan_Amount_Term'].mean(),inplace=True)
loan.isnull().sum()
```

You can also drop missing values with loan.dropna()

```python
#Understanding Dropping

# Drop columns and/or rows of MultiIndex DataFrame

midx = pd.MultiIndex(levels=[['deer', 'dog', 'eagle'],
                             ['speed', 'weight', 'length']],
                      codes=[[0, 0, 0, 1, 1, 1, 2, 2, 2],
                            [0, 1, 2, 0, 1, 2, 0, 1, 2]])
df = pd.DataFrame(index=midx, columns=['big', 'small'],
                   data=[[60, 35], [30, 12], [5.5, 1], [40, 20],
                         [25, 10], [4.5, 0.8], [320, 220],
                         [2, 0.8], [2.2, 0.2]])
df
```

```python
df.drop(index='dog', columns='small')
```

```python
df.drop(index='length',level=1)#For MultiIndex, level from which the labels will be removed.
```

```python
temp1 = loan['Credit_History'].value_counts(ascending=True)
temp2 = loan.pivot_table(values='Loan_Status',index=['Credit_History'],aggfunc=lambda x: x.map({'Y':1,'N':0}).mean())
print ('Frequency Table for Credit History:') 
print (temp1)

print ('\nProbility of getting loan for each Credit History class:')
print (temp2)
```

```python
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(121)
ax1.set_xlabel('Credit_History')
ax1.set_ylabel('Count of Applicants')
ax1.set_title("Applicants by Credit_History")
temp1.plot(kind='bar')

ax2 = fig.add_subplot(122)
temp2.plot(kind = 'bar')
ax2.set_xlabel('Credit_History')
ax2.set_ylabel('Probability of getting loan')
ax2.set_title("Probability of getting loan by credit history")
```
